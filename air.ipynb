```python
# Padronizar a coluna 'duration' para minutos
df['duration_minutes'] = (df['duration'].astype(float) * 60).round(2)
df['duration_minutes'].head(10)
```

```python
df.head(10).reset_index(drop=True)
```

```python
# Agrupar por rota e calcular a m√©dia de atrasos
df['route'] = df['source_city'] + ' - ' + df['destination_city']

media_desvio = df.groupby('route')['duration_minutes'].agg(['mean', 'std']).reset_index().round(2)
media_desvio_columns = ['route', 'mean_delay', 'std_delay']
df = df.merge(media_desvio, on='route', how='left')

df.head(10).reset_index(drop=True)
```

```python
# Verificar se o voo est√° atrasado
# Considerando um voo atrasado se a dura√ß√£o for maior que a m√©dia + desvio padr√£o
df['is_delayed'] = df['duration_minutes'] > df['mean'] + df['std']

df.head(10).reset_index(drop=True)
```

Classifica√ß√£o de desempenho por rota

```python
#Agrupando por companhia a√©rea
desempenho = df.groupby('airline').agg(
    total_voos=('airline', 'count'),
    total_atrasos=('is_delayed', 'sum'),
    pre√ßo_m√©dio=('price', 'mean'),
    dura√ß√£o_m√©dia=('duration_minutes', 'mean')
).reset_index()

#Calculo de percentual de atrasos
desempenho['percentual_atrasos'] = (desempenho['total_atrasos'] / desempenho['total_voos']) * 100

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

#Normalizando os dados
desempenho_normalizado = desempenho[['percentual_atrasos', 'pre√ßo_m√©dio', 'dura√ß√£o_m√©dia']].copy()
desempenho_normalizado = pd.DataFrame(scaler.fit_transform(desempenho_normalizado),
                                      columns=['atraso_normalizado', 'preco_normalizado', 'duracao_normalizada']
                                      , index=desempenho.index)


#Score final
desempenho_normalizado['score'] = (
    0.4 * desempenho_normalizado['atraso_normalizado'] +
    0.3 * desempenho_normalizado['preco_normalizado'] +
    0.3 * desempenho_normalizado['duracao_normalizada']
)

#Classificando
desempenho_normalizado['classificacao'] = pd.qcut(desempenho_normalizado['score'], q=5, labels=['E', 'D', 'C', 'B', 'A'])

#Jutando com os dados originais
desempenho_final = desempenho.join(desempenho_normalizado)

#Exibindo o resultado
print("\nClassifica√ß√£o de desempenho por companhia a√©rea:")
print("")
print(desempenho_final.sort_values('classificacao').reset_index(drop=True))
```

```python
#Heatmap de desempenho
plt.figure(figsize=(12, 8))
sns.heatmap(desempenho_final[['total_voos', 'total_atrasos', 'pre√ßo_m√©dio', 'dura√ß√£o_m√©dia', 'percentual_atrasos']].corr(), annot=True, cmap='coolwarm')
plt.title('Heatmap de Desempenho por Companhia A√©rea')
plt.xlabel('Vari√°veis')
plt.ylabel('Vari√°veis')
plt.tight_layout()
plt.show()
```

Veremos quais rotas cada empresa possue o melhor desempenho

```python
# Melhores rotas por companhia a√©rea
melhores_rotas_por_companhia = (
    desempenho_final.groupby(['airline', 'route'], as_index=False)
    .agg(
        media_score=('score', 'mean'),
        classificacao_mais_comum=('classificacao', lambda x: x.mode()[0]),
        total_voos=('airline', 'count')
    )
    .sort_values(by='media_score', ascending=False)
    .groupby('airline')
    .first()
    .reset_index()
)

#Exibindo resultados
melhores_rotas_por_companhia[['airline', 'route', 'media_score', 'classificacao_mais_comum', 'total_voos']]
```

Modelo de previs√£o de atraso

```python
# Modelo de previs√£o com regress√£o log√≠stica
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

#Selecionando colunas relevantes
colunas_relevantes = ['airline', 'source_city', 'destination_city', 'departure_time', 'arrival_time', 'class', 'days_left', 'price', 'duration_minutes', 'route', 'mean', 'std','stops']

target = 'is_delayed'

x = df[colunas_relevantes]
y = df[target].astype(int) # Convertendo booleano para int

#Definindo colunas num√©ricas e categ√≥ricas
categorical_cols = ['airline', 'source_city', 'destination_city', 'departure_time', 'stops', 'class', 'route']
numerical_cols = ['days_left', 'price', 'duration_minutes', 'mean', 'std']

#Preprocessamento
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_cols),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols)
    ])

#Pipeline
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression(max_iter=1000))
])

#Dividindo em treino e teste
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

#Treinando o modelo
pipeline.fit(x_train, y_train)

#Fazendo previs√µes
y_pred = pipeline.predict(x_test)

#Avalia√ß√£o do modelo
print("Matrix de Confus√£o:")
print(confusion_matrix(y_test, y_pred))
print("\nRelat√≥rio de Classifica√ß√£o:")
print(classification_report(y_test, y_pred))

#Acur√°cia do modelo
accuracy = pipeline.score(x_test, y_test)
print(f"\nAcur√°cia do modelo: {accuracy:.2f}")
```

```python
#Visualizando as previs√µes graficamente

plt.figure(figsize=(12, 6))
ax = sns.countplot(x=y_test, hue=y_pred, palette='pastel',legend=False,stat='probability')
for p in ax.patches:
    value = p.get_height() * 100
    ax.annotate(f'{value:.1f}%', (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='bottom', fontsize=12, color='black')   
plt.title('Compara√ß√£o entre Valores Reais e Previs√µes de Atrasos')
plt.xlabel('Valores Reais')
plt.ylabel('Previs√µes de Atrasos')
plt.xticks([0, 1], ['N√£o Atrasado', 'Atrasado'])
plt.legend(title='Previs√µes', loc='upper right', labels=['N√£o Atrasado', 'Atrasado'])
plt.tight_layout()
```

```python
#Verificando o Overfitting

from sklearn.metrics import classification_report

#Previs√µes de treino 
y_pred_train = pipeline.predict(x_train)
print("Teste:")
print(classification_report(y_train,y_pred_train))

#Previs√µes no teste
y_pred_test = pipeline.predict(x_test)
print("Teste:")
print(classification_report(y_test,y_pred_test))
```

```python
#Cross validation

from sklearn.model_selection import cross_val_score

scores = cross_val_score(pipeline, x, y, cv=5,scoring='accuracy')
print("Acur√°cia:",scores)
print("M√©dia:",np.mean(scores))

print("\n\n As m√©dias n√£o variaram muito, mostrando que o modelo possui uma boa generaliza√ß√£o.")
```

```python
#Curva de aprendizado
from sklearn.model_selection import learning_curve

train_size, train_scores, test_scores = learning_curve(pipeline, x, y, cv=5, scoring="accuracy")

plt.plot(train_size, np.mean(train_scores, axis=1), label = "Treino")
plt.plot(train_size, np.mean(test_scores, axis=1), label = "Teste")
plt.xlabel("Tamanho do conjunto de treino")
plt.ylabel("Acur√°cua")
plt.legend()
```

Ap√≥s toda a an√°lise, utilizando o modelo de regress√£o linear, √© poss√≠vel concluir que os dados s√£o suficientes para criar modelos de previs√£o com alta taxa de acur√°cidade. O desempenho quase perfeito mostra que h√° possibilidade de generalizar bem os dados mesmo quando ainda n√£o vistos. O modelo capta padr√µes genu√≠nos e n√£o apenas informa√ß√µes diretas sobre atraso.
